```python
#imported required libraries

import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
plt.rcParams['figure.figsize'] = (6.0, 6.0)
plt.rcParams['image.cmap'] = 'gray'
```

```python
#Read the image as an input data
sig=cv2.imread('/content/Hand_writtern.jpg')

#Pre-processing step
sig_c=sig[800:2100,500:3500] #cropped
sig_r=cv2.cvtColor(sig_c,cv2.COLOR_BGR2GRAY)  #Conversion

#Forming mask with the help of thresholding theory
reteval,sig_rt=cv2.threshold(sig_r,150,255,cv2.THRESH_BINARY_INV) #Created mask

#Display of the result of thresholding method
plt.imshow(sig_rt)
```

```python
#Post-processing step
b,g,r=cv2.split(sig_c)   #Spliting the oriogional image's channel
new=[b,g,r,sig_rt]   #added mask (as 4th channel) into original image shape

#A new image with 4 channels (includes an alpha with B G R)
sig_new=cv2.merge(new,4)

#Display an input (Hand writtern text) and output image (E-text)
plt.figure(figsize=[10,10])
plt.subplot(121);plt.imshow(sig_c);plt.title('Hand writtern signature')
plt.subplot(122);plt.imshow(sig_new);plt.title('E-signature');
```

```python
#Save the output image
cv2.imwrite("E-text.jpg",sig_new)
```
